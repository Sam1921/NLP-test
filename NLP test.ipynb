```python
from tqdm import tqdm
```

```python
def parse_reviews_in_a_column(review_column):
    """
    parse the reviews from reviews column of df in nlp through this function
    """
    parsed_review_list = []
    for review in tqdm(review_column):
        parsed_review_list.append(nlp(review))
    return parsed_review_list
```

```python
parsed_review_list = parse_reviews_in_a_column(df['reviews.text'])
```

```python
def get_sum_senti_score(doc):
    #Get ajectives from document
    adj_list =[] #Take a blank list
    for token in doc: #Iterate over each token in the documnet
        if token.pos_ == 'ADJ': #if the pos of the token is adjective
            if token.text in adj_count_df_filtered["index"].values: #if the token is in adjective dataframe
                adj_list.append(token.text) #take the adjective in the list
    
    #Get sentiment score for each adjective
    sentiment_list = [] #take a blank list
    for word in adj_list: #for each word in adjective list
        #get senntiment score from adjective dataframe and put ion the sentiment list
        sentiment_list.append(adj_count_df_filtered[adj_count_df_filtered["index"]==word]["Sentiment_Score"].values[0]) 
    
    #Return Sum of all sentiment score
    return sum(sentiment_list)
```

```python
sum_sentiment_list = [] #create a blank list
for doc in tqdm(parsed_review_list): #for each document in parsed document list
    sum_sentiment_list.append(get_sum_senti_score(doc)) #get the summed sentiment score and put ion the list
```

```python
final_df = pd.DataFrame({"Text":parsed_review_list,"Sentiment_Score":sum_sentiment_list})
```

```python
final_df["Text"] = final_df["Text"].apply(lambda doc: doc.text)
```

```python
"""
-2.4<=senti_score<-1 = 1
-1<senti_score<0 = 2
0 = 3
0
0<senti_score<0.8 = 4
0.8<senti_score<5.8 = 5
"""
```

```python
final_df.loc[(final_df["Sentiment_Score"]<1),"Rating"] = 1
final_df.loc[(final_df["Sentiment_Score"]>-1)&(final_df["Sentiment_Score"]<0),"Rating"] = 2
final_df.loc[(final_df["Sentiment_Score"]==0),"Rating"] = 3
final_df.loc[(final_df["Sentiment_Score"]>0)&(final_df["Sentiment_Score"]<0.8),"Rating"] = 4
final_df.loc[(final_df["Sentiment_Score"]>0.8),"Rating"] = 5
```
